```python
#1. We will start with removing outliers. 
#So far, we have discussed different methods to remove outliers. 
#Use the one you feel more comfortable with, define a function for that. Use the function to remove the outliers and apply it to the dataframe.

numerical = pd.read_csv('numerical.csv')
numerical.head()
```

```python
def plot_outliers(numerical):
    
    for col in numerical.columns:
        plt.figure()
        plt.title(col)
        numerical[col].plot(kind='box')

plot_outliers(numerical)
```

```python
data = pd.concat([categorical,numerical], axis=1)
```

```python
data.isna().sum()
```

```python
cols_outliers=["customer_lifetime_value","monthly_premium_auto"]
```

```python
#work on data because it will have impact on it as well, then create cols_outliers and apply only tho those which have outliers 
def remove_out(data,cols_outliers):
    for col in cols_outliers:
        iqr = np.percentile(data[col],75)-np.percentile(data[col],25)
        upper_limit = np.percentile(data[col],75) + 1.5*iqr
        data=data[data[col]<upper_limit]
    return data
```

```python
data = remove_out(data,cols_outliers).reset_index()
```

```python
#2. Create a copy of the dataframe for the data wrangling.
datacopy = data
```

```python
#3. Normalize the continuous variables. You can use any one method you want.
import numpy as np 
from sklearn.preprocessing import Normalizer

X = datacopy.select_dtypes(include = np.number)
X.head()
```

```python
transformer = Normalizer().fit(X)
x_normalized =  transformer.transform(X)
x = pd.DataFrame(x_normalized)
x.columns = X.columns
x.head()
```

```python
#4. Encode the categorical variables
categorical = datacopy.select_dtypes(include=['object', 'bool'])
categorical.head(1)
```

```python
#5. The time variable can be useful. Try to transform its data into a useful one. Hint: Day week and month as integers might be useful.

from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Load data
df = datacopy.select_dtypes(include=['object', 'bool'])

# One hot encoding for state
state_encoder = OneHotEncoder()
state_encoded = state_encoder.fit_transform(df[["state"]])
state_encoded_df = pd.DataFrame(state_encoded.toarray(), columns=state_encoder.get_feature_names(["state"]))
```

```python
# Ordinal encoding for coverage, employmentstatus, and location_code
ordinal_encoder = OrdinalEncoder(categories=[["Basic", "Extended", "Premium"], ["Unemployed", "Employed", "Medical Leave", "Disabled", "Retired"], ["Rural", "Suburban", "Urban"]])
ordinal_encoded = ordinal_encoder.fit_transform(df[["coverage", "employmentstatus", "location_code"]])
ordinal_encoded_df = pd.DataFrame(ordinal_encoded, columns=["coverage_encoded", "employmentstatus_encoded", "location_code_encoded"])
```

```python
# One hot encoding for marital_status, policy_type, renew_offer_type, sales_channel, and vehicle_class
one_hot_encoder = OneHotEncoder()
one_hot_encoded = one_hot_encoder.fit_transform(df[["marital_status", "policy_type", "renew_offer_type", "sales_channel", "vehicle_class"]])
one_hot_encoded_df = pd.DataFrame(one_hot_encoded.toarray(), columns=one_hot_encoder.get_feature_names(["marital_status", "policy_type", "renew_offer_type", "sales_channel", "vehicle_class"]))
```

```python
# Ordinal encoding for vehicle_size
vehicle_size_encoder = OrdinalEncoder(categories=[["Small", "Medsize", "Large"]])
vehicle_size_encoded = vehicle_size_encoder.fit_transform(df[["vehicle_size"]])
vehicle_size_encoded_df = pd.DataFrame(vehicle_size_encoded, columns=["vehicle_size_encoded"])

```

```python
def extract_month_year(df, effective_to_date):
    # Convert date column to datetime type
    df[effective_to_date] = pd.to_datetime(df[effective_to_date])
    
    # Extract month and year to new columns
    df["month"] = df[effective_to_date].dt.month
    df["year"] = df[effective_to_date].dt.year
    
    return df
df = datacopy
df = extract_month_year(df, "effective_to_date")
df.head()
```

```python
encoded_month_year = df[["month", "year"]]
encoded_month_year.head()
```

```python
# Concatenate all encoded dataframes
encoded_df = pd.concat([state_encoded_df, ordinal_encoded_df, one_hot_encoded_df, vehicle_size_encoded_df, encoded_month_year, x], axis=1)
encoded_df.head()
```
