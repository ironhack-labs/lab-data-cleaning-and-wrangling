```python
# describe Data to see difference between continuous and dsicrete Data 

'''Continuous Data:

Definition: Continuous data can take any value within a given range and can be measured with great precision.
Examples: Height, weight, temperature, time, distance, age.'''

'''Discrete Data:

Definition: Discrete data can only take specific, distinct values and cannot be measured with unlimited precision.
Examples: Number of siblings, number of cars in a parking lot, number of students in a class.'''

Numerical_df.describe().T
```

```python
# checking the number of unique value 
unique_values_list = []  

for column in Numerical_df:
    unique_values = Numerical_df[column].unique()
    unique_values_list.append((column, unique_values))

# Now unique_values_list contains tuples with column name and unique values
# You can access the unique values for each column as needed
for column_name, unique_values in unique_values_list:
    print(f'Unique values in {column_name}:')
    print(unique_values)
    print()
```

```python




# Create a copy of the original DataFrame to keep the original data
df_no_outliers = Numerical_df.drop(columns= 'number_of_open_complaints')

# Loop through each column in the DataFrame
for column in df_no_outliers.columns:
    Q1 = df_no_outliers[column].quantile(0.25)
    Q3 = df_no_outliers[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Create a mask to identify outliers
    outlier_mask = (df_no_outliers[column] < lower_bound) | (df_no_outliers[column] > upper_bound)

    # Remove outliers from the column
    df_no_outliers[column] = Numerical_df[column][~outlier_mask]

df_no_outliers
```

```python
df_no_outliers['number_of_open_complaints'] = Numerical_df['number_of_open_complaints']
```

```python
# creating a copy of the Data Numerical without outliers

df_Num_out = df_no_outliers.copy()
```

```python
 # separating between continuous Data and discrete Data

Continuous_Data = []

Discrete_Data = []

len_Row = Numerical_df.shape[0]


for column in df_Num_out:
    unique_values = df_Num_out[column].unique()
    if len(unique_values) > 0.1 * len_Row :
        Continuous_Data.append((column))
    else :
        Discrete_Data.append((column))


# creating a new Datav frame continuois and concreted Data

Discrete_df = df_Num_out[Discrete_Data]



Continiuous_df = df_Num_out[Continuous_Data]

```

```python
# creating Data Frame with only object

categorical_df = customer_df.select_dtypes(include = object) #or  exclude = 'number' or object

categorical_df
```

```python
# then will process to encoding 

# first will change object for gender and response to numerical as there is only two unique value 

categorical_df['response'] = np.where(categorical_df['response'] == 'Yes',1,0)

categorical_df['gender'] = np.where(categorical_df['gender'] == 'M',1,0)

```

```python
categorical_df
```

```python
# Both columns have the same information almost so we could combine them and keep only the useful info
# mapping policy type columns to have numerical value 

categorical_df['policy_type'] = categorical_df['policy_type'].str.replace('Auto','')


categorical_df

```

```python
# dropping the extra Policy column

categorical_df = categorical_df.drop(columns='policy')
```

```python
# checking the number of unique value 
unique_values_list = []  

for column in categorical_df:
    unique_values = categorical_df[column].unique()
    unique_values_list.append((column, unique_values))

# Now unique_values_list contains tuples with column name and unique values
# You can access the unique values for each column as needed
for column_name, unique_values in unique_values_list:
    print(f'Unique values in {column_name}:')
    print(unique_values)
    print()
```

```python
# turning all categorical data using mapping with numerical number for category with more than 2 options 
# Ordinal to coverage ok
# Ordinal to employmentstatus ok 
# Ordinal to location code ok
# Ordinal vehicle size ok

# using re+place instead of mapping as I could not make work the map function


categorical_df['employmentstatus'] = categorical_df['employmentstatus'].replace({'Unemployed' : 0,'Employed': 1 , 'Medical Leave': 2 ,'Disabled':3 ,'Retired': 4 })

categorical_df['location_code'] = categorical_df['location_code'].replace({'Rural': 0, 'Suburban':1, 'Urban':2})

categorical_df["coverage"] = categorical_df["coverage"].replace({"Basic" : 0, "Extended" : 1, "Premium" : 2})

categorical_df['education'] = categorical_df['education'].replace({'College': 0,'Bachelor': 1, 'High School or Below': 2 ,'Master': 3 ,'Doctor': 4})

categorical_df['vehicle_size'] = categorical_df['vehicle_size'].replace({'Small': 0,'Medsize': 1 ,'Large': 2})




# given that column "coverage" in the dataframe "data" has three categories:
# "basic", "extended", and "premium" and values are to be represented in the same order.


```
