```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Getting teh data
df_num = pd.read_csv(r'C:\Users\ljant\Desktop\Ironhack\10_KNN\lab-data-cleaning-and-wrangling\numerical.csv')

def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df_num)
```

```python
df_cat = pd.read_csv(r'C:\Users\ljant\Desktop\Ironhack\10_KNN\lab-data-cleaning-and-wrangling\categorical.csv')

def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df_cat)
```

```python
# Concatting
df = pd.concat([df_cat, df_num], axis = 1).reset_index(drop = True)
df
```

```python
###1. Removing outliers
'''
Only continious columns show outliers --> investigating/dropping in 'total_claim_amoiunt' and 'income'
'''
def outliers_iqr(data, columns, threshold=1.5):
    df_adj = df.copy()
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        df_adj = df_adj[(df_adj[column] >= lower_bound) & (df_adj[column] <= upper_bound)]

    return df_adj


clean_data = outliers_iqr(df, ['total_claim_amount', 'income'], threshold=1.5).reset_index(drop = True)
clean_data
```

```python
###2. Copy of the df for data wrangling
clean_df = clean_data.copy()
clean_df
```

```python
###3. Normalising the continious columns
# Get numeric df
def column_types(data):
    num_col = data.select_dtypes(include=['int', 'float', 'datetime64']).columns
    cat_col = data.select_dtypes(include=['object']).columns
    return num_col, cat_col

numerical_columns, categorical_columns = column_types(clean_df)

numerical_df = clean_df[numerical_columns]
categorical_df = clean_df[categorical_columns]
```

```python
# Differentiate between continious and discrete values (function)

def dicision(data, threshold=250):
    continuous_cols = []
    discrete_cols = []
    for column in data.columns:
        unique_count = data[column].nunique()
        if unique_count <= threshold:
            discrete_cols.append(column)
        else:
            continuous_cols.append(column)
    return continuous_cols, discrete_cols

continuous_columns, discrete_columns = dicision(numerical_df)
print("Continuous columns:")
print(continuous_columns)
print("\nDiscrete columns:")
print(discrete_columns)

continuous_df = numerical_df[continuous_columns]
discrete_df = numerical_df[discrete_columns]
```

```python
# Normalisations of continious columns

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
normalised_data = scaler.fit_transform(continuous_df)
normalised_data = pd.DataFrame(normalised_data, columns = continuous_df.columns)


df_num = pd.concat([discrete_df, normalised_data], axis = 1).reset_index(drop = True)
df_clean = pd.concat([df_num, categorical_df], axis = 1).reset_index(drop = True)
df_clean
```

```python
###4. Transforming Date into useful information
df_clean['effective_to_date'] = pd.to_datetime(df_clean['effective_to_date'])
data_types = df_clean.dtypes
print (data_types)
pd.set_option('display.max_rows', None)
```

```python
pd.reset_option('display.max_rows')
df_clean['effective_day'] = df_clean['effective_to_date'].dt.day
df_clean['effective_month'] = df_clean['effective_to_date'].dt.month

df_clean = df_clean.drop(['effective_to_date'], axis =1)
df_clean
#year is always 2011
```

```python
# As a function:
def check_numeric(dataframe):
    non_numeric = dataframe.select_dtypes(exclude=['int', 'float'])
    if non_numeric.empty:
        return []
    else:
        return non_numeric.columns.tolist()
    
non_numeric = check_numeric(df_clean)
if non_numeric:
    print("Non-numeric columns found:")
    for column in non_numeric:
        print(column)
else:
    print("All columns are numeric.")
```
