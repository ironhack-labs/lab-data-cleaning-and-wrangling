```python
# Removing outliers. Define a function for that 
# Use the function to remove the outliers and apply it to the dataframe


def remove_outliers (df, threshold=1.5):
    for column in df.columns:
        for value in df[column]:
    
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            lower = Q1 - threshold * IQR
            upper = Q3 + threshold * IQR

        df_removed = df[(df[column] >= lower) & (df[column] <= upper)]

        return df_removed

numerical_df = remove_outliers(numerical_df)


# Create a copy of the dataframe for the data wrangling

df_wrangling = numerical_df
```

```python
# Encode the categorical variables

# one hot encoding

hot_encode = ['state', 'marital_status', 'policy_type', 'policy', 'policy_type', 'renew_offer_type', 'sales_channel', 'vehicle_class']
df_encoded = pd.get_dummies(categorical_df, columns=hot_encode)
```

```python
# ordinal encoding

df_encoded["coverage"] = df_encoded["coverage"].map({"Basic" : 0, "Extended" : 1, "Premium" : 2})
df_encoded["vehicle_size"] = df_encoded["vehicle_size"].map({"Small" : 0, "Medsize" : 1, "Large" : 2})
df_encoded["employmentstatus"] = df_encoded["employmentstatus"].map({"Unemployed" : 0, "Disabled" : 1, "Medical Leave" : 2, "Retired" : 3, "Employed" : 4})
df_encoded['location_code'] = df_encoded['location_code'].map({"Rural" : 0, "Suburban" : 1, "Urban" : 2})
```

```python
# The time variable can be useful. 
# Try to transform its data into a useful one. Hint: Day week and month as integers might be useful.

df_encoded['effective_to_date'] = pd.to_datetime(df_encoded['effective_to_date'])
df_encoded['day_of_week'] = df_encoded['effective_to_date'].dt.dayofweek
df_encoded['month'] = df_encoded['effective_to_date'].dt.month
```
